```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the results file
file_path = "data/SSA_finetune_results.csv"
df = pd.read_csv(file_path)

# Extract measured and expected Fano Factors
measured_ff = df["fano_m"]
expected_ff = df["expected_fano_m_exact"]

# Plot measured vs. expected Fano Factor
plt.figure(figsize=(8, 6))
plt.scatter(expected_ff, measured_ff, alpha=0.7, label="Parameter Sets")
plt.plot([min(expected_ff), max(expected_ff)], [min(expected_ff), max(expected_ff)], 'r--', label="y = x")

# Labels and Title
plt.xlabel("Expected Fano Factor")
plt.ylabel("Measured Fano Factor")
plt.title("Measured vs. Expected Fano Factor")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
